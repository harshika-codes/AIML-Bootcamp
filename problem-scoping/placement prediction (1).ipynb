📊 Generating the Synthetic data
</h2>

```python
import numpy as np                        #num- numerical    py-python
import pandas as pd                       # as- alias
np.random.seed(15)
n_student = 1000

data = {   
    #student features
    "student_ID" : [f"User{i:03d}" for i in range(1, n_student + 1)],
    "Gender" : np.random.choice(["Male","Female"], n_student),
    "Age" : np.random.randint(20,22, n_student),
    "Department" : np.random.choice(["CSE","ECE","ME"], n_student),
    "Previous_academic_background" : np.random.choice(["A","B","C"], n_student),
    "Family_Income" : np.random.randint(10000,280000, n_student),

    #Academic features
    "Attendence_percentage" : np.clip(np.random.normal(75,15,n_student),20,100).round(1),
    "Semester_GPA " : np.clip(np.random.normal(7.0, 1.5, n_student),0, 10).round(2),                 
    "12th_percentage" :  np.clip(np.random.normal(75, 12, n_student), 0, 100),
    "Backlogs" : np.random.poisson(0.5,n_student),                                                 
    "certification_count" : np.random.randint(0,11, n_student),

    #skill feature
    "Aptitude_Test_Score" : np.clip(np.random.normal(75,15, n_student),0,100).round(2),               
    "Coding_Test_Score " : np.clip(np.random.normal(65,12, n_student),0,100).round(0),                               
    "Technical_Assessment_Score " : np.clip(np.random.normal(55,12, n_student),0,80).round(1),    
   
    #Experience Features
    "Internship_Count " : np.random.randint(0,6, n_student),
    "Project_Count " : np.random.randint(0,6, n_student),
    "Workshop_Participation " : np.random.randint(0,11, n_student),
    "Hackathon_Participation" : np.random.choice([0,1], n_student, p=[0.8,0.2]), 

    #Behavioural Features:
    "academic_warnings" : np.random.choice([0,1,2,3], n_student, p=[0.7, 0.2,0.08,0.02])
}
df = pd.DataFrame(data)
#print(data)
#print(df)
df
```

📊 Calculating the risk score
</h2>

```python
risk_score = (
    0.30 * (df["Attendence_percentage"] < 60).astype(int) +
    0.15 * (df["Semester_GPA "] < 7.5).astype(int) +  # Trailing space included
    0.10 * (df["Aptitude_Test_Score"] > 70).astype(int) +
    0.10 * (df["academic_warnings"] > 0).astype(int) +
    0.10 * (df["Project_Count "] > 3).astype(int) +  # Added trailing space
    0.10 * (df["Backlogs"] >= 4).astype(int) +
    0.05 * (df["Technical_Assessment_Score "] > 55).astype(int) +  # Added trailing space
    0.05 * (df["Internship_Count "] > 3).astype(int) +  # Added trailing space
    0.10 * (df["Workshop_Participation "] > 5).astype(int) +  # Added trailing space
    0.05 * (df["Hackathon_Participation"] > 0).astype(int)
)

risk_score += np.random.normal(0, 0.05, len(df))
```

<h2 style="

📊 Printing Table
</h2>

```python
df["Placement"] = (risk_score >= 0.50).astype(int)
df
```
