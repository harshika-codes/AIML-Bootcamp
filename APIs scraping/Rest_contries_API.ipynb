```python
import requests
import pandas as pd
import json
```

```python
url = "https://countriesnow.space/api/v0.1/countries/capital"

response = requests.get(url)

data = response.json()

countries = data["data"]

df = pd.DataFrame(countries)

df.head(20)
```

```python
df.to_csv("Countries.csv", index=False)
print("CSV Saved Successfully!")
```
