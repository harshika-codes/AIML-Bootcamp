```python
import requests
import pandas as pd

# API URLd
url = "https://jsonplaceholder.typicode.com/posts"

# API request
response = requests.get(url)

# JSON data
data = response.json()

# DataFrame
df = pd.DataFrame(data)

# Show table
df
```
