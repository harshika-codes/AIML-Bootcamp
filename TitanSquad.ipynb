## ⚙️ Process of Web Scraping on Hacker News

### 1️⃣ Sending the HTTP Request
The scraper sends an HTTP GET request to the Hacker News website using the Requests library to access the webpage content.

### 2️⃣ Retrieving Web Page Content
The Hacker News server processes the request and returns the raw HTML source code containing news articles and related information.

### 3️⃣ Parsing the HTML Content
The received HTML is parsed using BeautifulSoup to create a structured and navigable document tree.

### 4️⃣ Extracting the Required Data
Relevant information such as:
- News Title
- Article Link
- Rank
- Points
- Author Name
- Number of Comments

is extracted using HTML tags and CSS selectors.

### 5️⃣ Storing the Data
The extracted data is organized into a Pandas DataFrame and can be exported to:
- CSV File
- Excel File
- JSON Format

for further analysis.

### 6️⃣ Data Analysis and Insights
The collected Hacker News data is analyzed to identify:
- Trending Topics
- Most Popular Articles
- User Engagement
- Story Rankings

which helps in understanding current technology and startup trends.

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
soup1 = soup.find_all('span', class_="titleline")
soup1
```

📊 fetching the Data
</div>

```python
soup1 =[]
```

```python
news = soup.find_all('tr', class_='athing')

for i in news:

    rank = i.find('span', class_='rank').text.strip()

    title = i.find('span', class_='titleline').get_text(strip=True)

    link = i.find('span', class_='titleline').find('a')['href']

    website = i.find('span', class_='sitestr')
    website = website.text if website else 'N/A'

    soup1.append({
        'Rank': rank,
        'Title': title,
        'Link': link,
        'Website': website
    })

soup1
```

```python
soup1
```

```python
df = pd.DataFrame(soup1)
df
```
