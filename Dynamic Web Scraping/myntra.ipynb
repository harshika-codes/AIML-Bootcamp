```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

<h3 style="
    background:linear-gradient(to right,#e0c3fc,#f9f9f9);
    color:#5a189a;
    padding:12px 20px;
    border-radius:30px;
    width:fit-content;
    font-size:26px;
    font-family:Arial, sans-serif;
    box-shadow:0 3px 8px rgba(0,0,0,0.1);
">
    📌 Chrome Setup
</h3>

```python
# -----------------------------
# 1. Chrome setup
# -----------------------------
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
wait = WebDriverWait(driver, 15)
```

<h3 style="
    background:linear-gradient(to right,#e0c3fc,#f9f9f9);
    color:#5a189a;
    padding:12px 20px;
    border-radius:30px;

    box-shadow:0 3px 8px rgba(0,0,0,0.1);
">
    📌 Search query
</h3>

```python
# -----------------------------
# 2. Search query
# -----------------------------
search_query = "shoes"   # yaha apna product likho
url = f"https://www.myntra.com/{search_query}"
driver.get(url)

time.sleep(5)
```

<h3 style="
    background:linear-gradient(to right,#e0c3fc,#f9f9f9);
    color:#5a189a;
    padding:12px 20px;
    border-radius:30px;
    width:fit-content;
    font-size:26px;
    font-family:Arial, sans-serif;
    box-shadow:0 3px 8px rgba(0,0,0,0.1);
">
    📌 Loading
</h3>

```python
# -----------------------------
# 3. Scroll a little to load products
# -----------------------------
driver.execute_script("window.scrollTo(0, 1200);")
time.sleep(3)
```

```python
# -----------------------------
# 4. Wait for products
# -----------------------------
wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "li.product-base")))

products = driver.find_elements(By.CSS_SELECTOR, "li.product-base")

data = []

for product in products:
    try:
        # Brand
        try:
            brand = product.find_element(By.CSS_SELECTOR, "h3.product-brand").text
        except:
            brand = "N/A"

        # Product title
        try:
            title = product.find_element(By.CSS_SELECTOR, "h4.product-product").text
        except:
            title = "N/A"

        # Price
        try:
            price = product.find_element(By.CSS_SELECTOR, "span.product-discountedPrice").text
        except:
            try:
                price = product.find_element(By.CSS_SELECTOR, "span.product-strike").text
            except:
                try:
                    price = product.find_element(By.CSS_SELECTOR, "div.product-price").text
                except:
                    price = "N/A"

        # Rating
        try:
            rating = product.find_element(By.CSS_SELECTOR, "div.product-ratingsContainer span").text
        except:
            rating = "N/A"

        # Product link
        try:
            link = product.find_element(By.TAG_NAME, "a").get_attribute("href")
        except:
            link = "N/A"

        data.append({
            "Brand": brand,
            "Product Title": title,
            "Price": price,
            "Rating": rating,
            "Product Link": link
        })

    except Exception as e:
        print("Error in one product:", e)
```

```python
# -----------------------------
# 5. Convert to DataFrame
# -----------------------------
df = pd.DataFrame(data)
print("Total products scraped:", len(df))
df.head(10)
```
