```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
```

">
    📌 Driver Set Up
</h3>

```python
# Chrome options
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")


driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

<h3 style="
    background:linear-gradient(to right,#e0c3fc,#f9f9f9);
    color:#5a189a;
    padding:12px 20px;
    border-radius:30px;
    width:fit-content;
    font-size:26px;
    font-family:Arial, sans-serif;
    box-shadow:0 3px 8px rgba(0,0,0,0.1);
">
    📌 Open Youtube And Search
</h3>

```python
search_query = "web scraping"

driver.get("https://www.youtube.com/")
time.sleep(3)

# Search box find karo
search_box = driver.find_element(By.NAME, "search_query")
search_box.send_keys(search_query)
search_box.send_keys(Keys.ENTER)

time.sleep(5)
```

```python
for i in range(5): 
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(3)
```

<h3 style="
    background:linear-gradient(to right,#e0c3fc,#f9f9f9);
    color:#5a189a;
    padding:12px 20px;
    border-radius:30px;

">
    📌 Scrape Vedio Data
</h3>

```python
videos = driver.find_elements(By.XPATH, '//*[@id="dismissible"]')

data = []

for video in videos:
    try:
        title = video.find_element(By.ID, "video-title").text
    except:
        title = "N/A"
    
    try:
        link = video.find_element(By.ID, "video-title").get_attribute("href")
    except:
        link = "N/A"
    
    try:
        channel = video.find_element(By.XPATH, './/*[@id="channel-name"]//a').text
    except:
        channel = "N/A"
    
    try:
        views = video.find_element(By.XPATH, './/*[@id="metadata-line"]/span[1]').text
    except:
        views = "N/A"
    
    try:
        upload_time = video.find_element(By.XPATH, './/*[@id="metadata-line"]/span[2]').text
    except:
        upload_time = "N/A"

    
    if title != "N/A" and title.strip() != "":
        data.append({
            "Title": title,
            "Channel": channel,
            "Views": views,
            "Uploaded": upload_time,
            "Link": link
        })
```
