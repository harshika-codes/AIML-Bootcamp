```python
import time
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

<h2 style="
    text-align:center;
    color:#5a189a;
    font-family:'Trebuchet MS', sans-serif;
    font-size:32px;
    background:#f8edff;
    padding:12px;
    border-left:8px solid #c77dff;
    border-radius:12px;
">
    Chrome Setup
</h2>

```python
# -----------------------------
# 1. Chrome setup
# -----------------------------
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
wait = WebDriverWait(driver, 15)
```

<h2 style="
    text-align:center;
    color:#5a189a;
    font-family:'Trebuchet MS', sans-serif;
    font-size:32px;
    background:#f8edff;
    padding:12px;
    border-left:8px solid #c77dff;
    border-radius:12px;
">
    Search Query
</h2>

```python
# -----------------------------
# 2. Search query
# -----------------------------
job_keyword = "data analyst"   # yaha apni job likho
location = "India"             # yaha location likho

url = f"https://in.indeed.com/jobs?q={job_keyword.replace(' ', '+')}&l={location.replace(' ', '+')}"
driver.get(url)

time.sleep(5)
```

<h2 style="
    text-align:center;
    color:#5a189a;
    font-family:'Trebuchet MS', sans-serif;
    font-size:32px;
    background:#f8edff;
    padding:12px;
    border-left:8px solid #c77dff;
    border-radius:12px;
">
    loading
</h2>

```python
# -----------------------------
# 3. Wait for job cards
# -----------------------------
wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "[data-testid='slider_item']")))

job_cards = driver.find_elements(By.CSS_SELECTOR, "[data-testid='slider_item']")

data = []

for job in job_cards:
    try:
        # Job title
        try:
            title = job.find_element(By.CSS_SELECTOR, "h2 a span").text
        except:
            title = "N/A"

        # Company
        try:
            company = job.find_element(By.CSS_SELECTOR, "[data-testid='company-name']").text
        except:
            company = "N/A"

        # Location
        try:
            job_location = job.find_element(By.CSS_SELECTOR, "[data-testid='text-location']").text
        except:
            job_location = "N/A"

        # Salary
        try:
            salary = job.find_element(By.CSS_SELECTOR, "[data-testid='attribute_snippet_testid']").text
        except:
            salary = "N/A"

        # Job link
        try:
            link = job.find_element(By.CSS_SELECTOR, "h2 a").get_attribute("href")
            if link and link.startswith("/"):
                link = "https://in.indeed.com" + link
        except:
            link = "N/A"

        data.append({
            "Job Title": title,
            "Company": company,
            "Location": job_location,
            "Salary": salary,
            "Job Link": link
        })

    except Exception as e:
        print("Error in one job:", e)
```

```python
# -----------------------------
# 4. Convert to DataFrame
# -----------------------------
df = pd.DataFrame(data)
print("Total jobs scraped:", len(df))
df.head(10)
```
