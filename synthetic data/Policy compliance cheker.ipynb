```python
import numpy as np
import pandas as pd

np.random.seed(54)

n_documents = 10000

data = {

    # Basic Information
    "Document_ID": [f"DOC{i:05d}" for i in range(1, n_documents+1)],

    "Organization_Type": np.random.choice(
        ["University","Company","Government","Hospital","Bank","NGO"],
        n_documents,
        p=[0.25,0.30,0.15,0.10,0.10,0.10]
    ),

    "Department": np.random.choice(
        ["HR","Finance","Legal","Academic","Research","Administration","Operations"],
        n_documents
    ),

    "Document_Type": np.random.choice(
        ["Project Report","Research Paper","Application",
         "Policy Document","Invoice","Legal Document",
         "Medical Report","Resume","Contract"],
        n_documents
    ),

    "Submitted_By": np.random.choice(
        ["Student","Employee","Faculty","Customer","Vendor","Citizen"],
        n_documents
    ),

    "Language": np.random.choice(
        ["English","Hindi","Other"],
        n_documents,
        p=[0.75,0.20,0.05]
    ),

    # Document Details
    "Total_Pages": np.random.randint(2,150,n_documents),

    "Mandatory_Sections_Present": np.random.choice(
        ["Yes","No"],
        n_documents,
        p=[0.85,0.15]
    ),

    "Missing_Sections_Count": np.random.poisson(1,n_documents),

    "Policy_Violations_Count": np.random.poisson(2,n_documents),

    # Quality Features
    "Formatting_Score": np.clip(
        np.random.normal(85,10,n_documents),40,100
    ).round(1),

    "Grammar_Score": np.clip(
        np.random.normal(82,12,n_documents),40,100
    ).round(1),

    "Plagiarism_Percentage": np.clip(
        np.random.normal(15,10,n_documents),0,60
    ).round(),

    "Document_Completeness": np.clip(
        np.random.normal(90,8,n_documents),40,100
    ).round(1),

    "Readability_Score": np.clip(
        np.random.normal(80,10,n_documents),40,100
    ).round(1),

    # Verification
    "Signature_Present": np.random.choice(
        ["Yes","No"],
        n_documents,
        p=[0.90,0.10]
    ),

    "Required_Attachments": np.random.choice(
        ["Yes","No"],
        n_documents,
        p=[0.88,0.12]
    ),

    "Submission_On_Time": np.random.choice(
        ["Yes","No"],
        n_documents,
        p=[0.80,0.20]
    ),

    "Policy_Version": np.random.choice(
        ["v1.0","v2.0","v3.0","v4.0"],
        n_documents
    ),

    "Sensitive_Data_Detected": np.random.choice(
        ["Yes","No"],
        n_documents,
        p=[0.08,0.92]
    )
}

df = pd.DataFrame(data)

df
```

```python
# Risk Score Calculation

df["Risk_Score"] = (

    # Missing Sections (Maximum Impact)
    df["Missing_Sections_Count"] * 8

    # Policy Violations
    + df["Policy_Violations_Count"] * 7

    # Plagiarism
    + df["Plagiarism_Percentage"] * 0.50

    # Poor Formatting
    + (100 - df["Formatting_Score"]) * 0.15

    # Poor Grammar
    + (100 - df["Grammar_Score"]) * 0.10

    # Incomplete Document
    + (100 - df["Document_Completeness"]) * 0.20

    # Poor Readability
    + (100 - df["Readability_Score"]) * 0.10

    # Mandatory Sections Missing
    + np.where(df["Mandatory_Sections_Present"] == "No", 12, 0)

    # Signature Missing
    + np.where(df["Signature_Present"] == "No", 8, 0)

    # Required Attachments Missing
    + np.where(df["Required_Attachments"] == "No", 8, 0)

    # Late Submission
    + np.where(df["Submission_On_Time"] == "No", 6, 0)

    # Sensitive Data Found
    + np.where(df["Sensitive_Data_Detected"] == "Yes", 10, 0)

)

# Keep score between 0 and 100
df["Risk_Score"] = np.clip(df["Risk_Score"], 0, 100).round(2)
```
