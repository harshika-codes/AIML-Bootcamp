📊 Generating the Synthetic data
</h2>

```python
import numpy as np
import pandas as pd

np.random.seed(15)
n_records = 1000

data = {
    # Basic sales/store info
    "Store_ID": [f"Store{i:03d}" for i in range(1, n_records + 1)],
    "Product_Category": np.random.choice(["Electronics", "Clothing", "Grocery", "Furniture"], n_records),
    "Region": np.random.choice(["North", "South", "East", "West"], n_records),
    "Sales_Channel": np.random.choice(["Online", "Offline"], n_records),

    # Product / pricing features
    "Product_Price": np.random.randint(200, 5000, n_records),
    "Discount_Percentage": np.random.randint(0, 51, n_records),
    "Competitor_Price": np.random.randint(150, 4800, n_records),

    # Marketing / promotion features
    "Marketing_Spend": np.random.randint(1000, 50000, n_records),
    "Ad_Clicks": np.random.randint(50, 5000, n_records),
    "Promotional_Offer": np.random.choice([0, 1], n_records, p=[0.6, 0.4]),

    # Time / seasonal features
    "Month": np.random.randint(1, 13, n_records),
    "Weekend_Sales": np.random.choice([0, 1], n_records, p=[0.4, 0.6]),
    "Holiday_Season": np.random.choice([0, 1], n_records, p=[0.7, 0.3]),

    # Inventory / customer behaviour
    "Stock_Available": np.random.randint(10, 500, n_records),
    "Customer_Rating": np.clip(np.random.normal(4.0, 0.7, n_records), 1, 5).round(1),
    "Return_Rate": np.clip(np.random.normal(8, 4, n_records), 0, 30).round(1),

    # Target variable
    "Units_Sold": np.random.randint(20, 500, n_records)
}

df_sales = pd.DataFrame(data)

df_sales
```

<h2 style="
background:#7B2CBF;

📊  Calculating the Risk Score
</h2>

```python
risk_score = (
    0.20 * (df_sales["Discount_Percentage"] < 10).astype(int) +
    0.15 * (df_sales["Marketing_Spend"] < 10000).astype(int) +
    0.15 * (df_sales["Customer_Rating"] < 3.5).astype(int) +
    0.10 * (df_sales["Return_Rate"] > 15).astype(int) +
    0.10 * (df_sales["Stock_Available"] > 300).astype(int) +
    0.10 * (df_sales["Product_Price"] > 3000).astype(int) +
    0.05 * (df_sales["Promotional_Offer"] == 0).astype(int) +
    0.05 * (df_sales["Holiday_Season"] == 0).astype(int) +
    0.05 * (df_sales["Weekend_Sales"] == 0).astype(int) +
    0.05 * (df_sales["Units_Sold"] < 80).astype(int)
)

df_sales["Risk_Score"] = risk_score.round(2)

df_sales
```

<h2 style="
background:#7B2CBF;
color:white;
padding:10px 18px;
border-radius:8px;
display:inline-block;
font-family:'Segoe UI',sans-serif;">
📊  Adding the Risk Levels
</h2>

```python
df_sales["Risk_Level"] = pd.cut(
    df_sales["Risk_Score"],
    bins=[-0.1, 0.30, 0.60, 1.0],
    labels=["Low", "Medium", "High"]
)

df_sales
```
