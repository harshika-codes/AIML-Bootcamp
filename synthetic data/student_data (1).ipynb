```python
import numpy as np #num- numerical    py-python
import pandas as pd # as- alias
np.random.seed(54)
n_student = 100

data = {
    "name" : np.random.choice(["Karan","Harsh","Nancy","Harshika"], n_student),
    "age" : np.random.randint(18, 25, n_student),
    "city" : np.random.choice(["Ambala","Radaur","Ladwa","Delhi"])
}
df = pd.DataFrame(data)
df
```

```python
import sqlite3 as sql
import pandas as pd
```

```python
# Create a cursor object using the connection
cursor = connection.cursor()
# cursor helps u to traverse over the records in a db
```

```python
connection.commit()
```

```python
student_data= list(df.itertuples(index=False , name = None))
cursor.executemany("INSERT INTO feilds2(name, age, city) VALUES (?,?,?)", student_data)
connection.commit()
```

```python
cursor.execute('''SELECT * FROM feilds2''')
```

```python
results = cursor.fetchall()
```
