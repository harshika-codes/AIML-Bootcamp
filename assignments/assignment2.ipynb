1. What is the difference between a list and a tuple, beyond "one is mutable and one isn't"? Why would you ever choose a tuple?</br>

2. Why is dictionary lookup O(1) on average? What data structure makes this possible internally?</br>
**Ans** : Dictionary lookup is O(1) on average because dictionaries use a hash table internally. The key's hash value directly determines where the data is stored.

3. What does "hashable" mean, and which built-in types are NOT hashable?</br>
**ANS** : A hashable object has a fixed hash value that does not change during its lifetime. Immutable types are usually hashable. Lists, dictionaries, and sets are not hashable.

5. Why is set membership testing faster than list membership testing?</br>
**Ans** : Sets use hash tables, giving O(1) average lookup time, while lists require checking elements one by one, giving O(n) time complexity.

**Ans** : No new list is created. Both variables point to the same list object in memory. Changes made through one variable affect the other.

**Bug 1 — Aliasing Trap**

This function is supposed to add a discount tag to a list of order</br>
WITHOUT modifying the caller's original list. It has a bug — find it.</br>
def apply_discount_tag(orders):</br>
 tagged = orders</br>
 tagged.append("DISCOUNT_APPLIED")</br>
 return tagged</br>
original_orders = ["order_1", "order_2"]</br>
new_orders = apply_discount_tag(original_orders)</br>
print("Original:", original_orders) # Did this change unexpectedly?</br>
print("New:", new_orders)</br>
**Fix ↓** 

```python
def apply_discount_tag(orders):
    tagged = orders.copy()   # new copy
    tagged.append("DISCOUNT_APPLIED")
    return tagged

original_orders = ["order_1", "order_2"]
new_orders = apply_discount_tag(original_orders)

print("Original:", original_orders)
print("New:", new_orders)
```

**Bug 6 — Unhashable Type**

This is meant to group transactions by (date, category) and accumulate</br>
totals into a dict keyed by that pair. It throws a TypeError. Why, and fix it?</br>
transactions = [</br>
 {"date": "2026-01-01", "category": ["food"], "amount": 200},</br>
 {"date": "2026-01-01", "category": ["food"], "amount": 150},</br>
]</br>
totals = {}</br>
for t in transactions:</br>
 key = (t["date"], t["category"]) # <-- problem</br>
 totals[key] = totals.get(key, 0) + t["amount"]</br>
print(totals)
**Fix :↓**

```python
transactions = [
    {"date": "2026-01-01", "category": ["food"], "amount": 200},
    {"date": "2026-01-01", "category": ["food"], "amount": 150},
]

totals = {}

for t in transactions:
    key = (t["date"], tuple(t["category"]))   # list → tuple
    totals[key] = totals.get(key, 0) + t["amount"]

print(totals)
```

```python
def top_scorer(names, scores):
    student_scores = dict(zip(names, scores))
    return max(student_scores, key=student_scores.get)

print(top_scorer(["Asha", "Ravi", "Meena"], [88, 95, 91]))
```

**6. Using collections.defaultdict, group a list of words by their first letter.** </br>

 pass</br>
print(group_by_first_letter(["apple", "ant", "banana", "cat", "car"]))</br>
**{'a': ['apple', 'ant'], 'b': ['banana'], 'c': ['cat', 'car']}** 

```python
from collections import defaultdict

def group_by_first_letter(words):
    grouped = defaultdict(list)

    for word in words:
        grouped[word[0]].append(word)

    return dict(grouped)

print(group_by_first_letter(["apple", "ant", "banana", "cat", "car"]))
```

(Hint: this is conceptually related to the Boyer-Moore majority vote idea —</br>
it won't be exact for all distributions, but explore the tradeoff.)* </br>
def streaming_mode_approx(stream):</br>
 pass

**ANSWER : Building a full Counter is not the most memory-efficient approach because it stores the frequency of every unique value in memory. For 200 million rows with many unique values, memory usage can become very large.</br>
Counter.most_common(1) gives an exact answer but requires storing all counts. A streaming max-tracking approach uses constant memory and is much more memory-efficient, but it may only provide an approximate answer and can miss the true most frequent item when the data distribution is complex.**

```python
def streaming_mode_approx(stream):
    candidate = None
    count = 0

    for item in stream:
        if count == 0:
            candidate = item
            count = 1
        elif item == candidate:
            count += 1
        else:
            count -= 1

    return candidate


data = [1, 2, 1, 1, 3, 1, 4]
print(streaming_mode_approx(data))
```

**Case:** "QuickCart" Flash-Sale Cart Service</br>
QuickCart runs a flash sale. Each cart event arrives as:</br>
{"cart_id": "C1", "product_id": "P10", "action": "add", "qty": 2}</br>
Within minutes, support tickets pour in: customers see duplicate items in their cart, totals are wrong, and the page is</br>
slow to load for carts with 50+ items.</br>
**Analyze** — what data structure choices likely caused each symptom?</br>
Duplicate items appearing → ?</br>
Wrong totals → ?</br>
Slow page loads for big carts → ?</br>
Design — propose a cart representation (which structure(s) would you use to store cart contents, keyed how?) that</br>
prevents all three issues.</br>
Implement — write a Cart class/functions using your design: add_item(product_id, qty) ,</br>
remove_item(product_id) , get_total(price_lookup) .</br>
**Your implementation:** </br>
class Cart:</br>
 def __init__(self):</br>
 pass</br>
 def add_item(self, product_id, qty):</br>
 pass</br>
 def remove_item(self, product_id):</br>
 pass</br>
 def get_total(self, price_lookup):</br>
 **price_lookup: dict of product_id -> price**</br>
 pass

**ANSWER : Duplicate items appearing → Using a list to store products may allow the same product ID to be added multiple times.</br>
Wrong totals → Duplicate entries or incorrect quantity tracking can cause total calculations to be inaccurate.</br>
Slow page loads for big carts → Using a list requires O(n) searches to find products. With 50+ items and frequent updates, performance degrades.** </br>
</br>
**Benefits:
Prevents duplicate products (unique keys)</br>
Keeps quantities accurate</br>
Fast O(1) lookup, update, and removal**</br>

```python
class Cart:

    def __init__(self):
        self.items = {}

    def add_item(self, product_id, qty):
        if product_id in self.items:
            self.items[product_id] += qty
        else:
            self.items[product_id] = qty

    def remove_item(self, product_id):
        self.items.pop(product_id, None)

    def get_total(self, price_lookup):
        total = 0

        for product_id, qty in self.items.items():
            total += price_lookup.get(product_id, 0) * qty

        return total


cart = Cart()

cart.add_item("P10", 2)
cart.add_item("P20", 1)
cart.add_item("P10", 3)

price_lookup = {
    "P10": 100,
    "P20": 50
}

print(cart.items)
print("Total:", cart.get_total(price_lookup))

cart.remove_item("P20")

print(cart.items)
print("Total after removal:", cart.get_total(price_lookup))
```
