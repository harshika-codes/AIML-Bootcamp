- web Scrapping is a powerful technique used to extract data from websites.
- Fetch the Web Page -> Parsing the content -> Mark the content to be extracted/retrieved -> Store it for further anylasis.
- commonly used for :</br>
  . Data collection</br>
  . Research and various other applications  # list all uses also??
  

**-How Web Scrapping is done ?**</br>
-sending a http request : send to target Web Server to retrieved to desired Web page</br>
-Retriving web page content : server responds with the web page content, usually HTML format.</br>
-Parsing the HTML : HTML content is parsed to locate the specific data elements using liberaries , tools designed for this purpose</br>
   [BeautifulSoup  , Scrapy]</br>
-extracting Data : Revelent data is extracted from the parsed HTML structure based on predefined rules or creteria.</br>
-Storing Data : The extracted data can be stored in various format such as CSV, json, or databases,for further anylasis or processing</br>
-Handling Challenges : Web Scrap[pers may need to address challenges like:</br>
 . Anti scrapping measures</br>
 . Dynamic content loading(Javascript)</br>
 . Legal considerations

```python
from bs4 import BeautifulSoup
import pandas as pd

# parsing the HTML content stored in html_doc
soup = BeautifulSoup(html_doc, 'html.parser')
soup
```

```python
print(soup.prettify())
```

```python
books = soup.find_all('div', class_ = 'book')
books
```

```python
book_data=[]
```

```python
for book in books:
    title = book.find('p', class_='title').get_text(strip = True)
    author =book.find('p',class_='author').get_text(strip = True)
    price =book.find('p',class_='price').get_text(strip = True)
    publication_year =book.find('p',class_='publication-year').get_text(strip = True)
    genre =book.find('p',class_='genre').get_text(strip = True)
    description = book.find('p',class_='description').get_text(strip = True)
    
    book_data.append({
        'Title' : title,
        'Author' : author,
        'Price' : price,
        'Publication_year' :publication_year,
        'Genre' : genre,
        'Description' : description
    })
book_data
```

```python
df = pd.DataFrame(book_data)
df
```

```python
df.to_csv("book_data", index = False)
```
