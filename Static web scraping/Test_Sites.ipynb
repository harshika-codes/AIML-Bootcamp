```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://webscraper.io/test-sites/e-commerce/static"

response = requests.get(url)

print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, "html.parser")
```

```python
products = soup.find_all("div", class_="thumbnail")

print(len(products))
```

```python
product_names = []
prices = []
descriptions = []
reviews = []

for product in products:

    # Product Name
    try:
        name = product.find("a", class_="title").get_text(strip=True)
    except:
        name = "N/A"

    # Price
    try:
        price = product.find("h4", class_="price").get_text(strip=True)
    except:
        price = "N/A"

    # Description
    try:
        desc = product.find("p", class_="description").get_text(strip=True)
    except:
        desc = "N/A"

    # Reviews
    try:
        review = product.find("p", class_="pull-right").get_text(strip=True)
    except:
        review = "N/A"

    product_names.append(name)
    prices.append(price)
    descriptions.append(desc)
    reviews.append(review)
```

```python
df = pd.DataFrame({
    "Product Name": product_names,
    "Price": prices,
    "Description": descriptions,
    "Reviews": reviews
})

df
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

headers = {
    "User-Agent": "Mozilla/5.0"
}

product_names = []
prices = []
descriptions = []
reviews = []
ratings = []

# Total 20 pages
for page in range(1, 21):

    url = f"https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page={page}"

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    products = soup.find_all("div", class_="thumbnail")

    print(f"Scraping Page {page}... Found {len(products)} Products")

    for product in products:

        # Product Name
        try:
            name = product.find("a", class_="title")["title"]
        except:
            name = "N/A"

        # Price
        try:
            price = product.find("h4", class_="price").text.strip()
        except:
            price = "N/A"

        # Description
        try:
            desc = product.find("p", class_="description").text.strip()
        except:
            desc = "N/A"

        # Reviews
        try:
            review = product.find("p", class_="pull-right").text.strip()
        except:
            review = "N/A"

        # Rating
        try:
            rating = len(product.find_all("span", class_="glyphicon-star"))
        except:
            rating = 0

        product_names.append(name)
        prices.append(price)
        descriptions.append(desc)
        reviews.append(review)
        ratings.append(rating)

df = pd.DataFrame({
    "Product Name": product_names,
    "Price": prices,
    "Description": descriptions,
    "Reviews": reviews,
    "Rating": ratings
})

print("Total Products:", len(df))

df
```
