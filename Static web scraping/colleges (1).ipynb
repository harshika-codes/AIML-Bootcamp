📌 Parsing The Content

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
print(soup.prettify())
```

<h3 style="

">
📌 Fetch The Data
</h3>

```python
college = soup.find_all('div', class_="col-12 px-1")
print(len(college))
```

```python
for i in college:
    print(i.text[:100])
```

```python
rows = soup.find_all("tr")
print(len(rows))
```

```python
 college=[]
```

```python
college = []

rows = soup.find_all("tr")

for row in rows:

    name = row.find("div", class_="clg-name-address")
    name = name.get_text(strip=True) if name else "N/A"

    fee = row.find("span", class_="text-green")
    fee = fee.get_text(strip=True) if fee else "N/A"

    review = row.find("span", class_="align-items-center")
    review = review.get_text(strip=True) if review else "N/A"

    location = row.find("div", class_="font-weight-medium")
    location = location.get_text(strip=True) if location else "N/A"

    if name != "N/A":
        college.append({
            "Name": name,
            "Course Fee": fee,
            "Review": review,
            "Location": location
        })

print(college)
```

```python
college
```

```python
df = pd.DataFrame(college)
df
```
